```python
import sys
import os

# Add src/ to the system path so we can import utils
sys.path.append(os.path.abspath("../src"))

from utils import load_data
```

```python
df = load_data("../data/customerdata.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


%matplotlib inline

# Quick look at the data
df.info()
df.describe()
df.head()
```

```python

# Check column names
print("Columns:", df.columns.tolist())
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Relationship between app usage and spend
sns.jointplot(data=df, x='Time on App', y='Yearly Amount Spent', kind='reg')
plt.show()
```

```python
# Relationship between website usage and spend
sns.jointplot(data=df, x='Time on Website', y='Yearly Amount Spent', kind='reg')
plt.show()
```

```python
# Relationship between membership length and spend
sns.jointplot(data=df, x='Length of Membership', y='Yearly Amount Spent', kind='reg')
plt.show()
```

```python

# Distribution of yearly spend
sns.histplot(df['Yearly Amount Spent'], kde=True, bins=30)
plt.title("Distribution of Yearly Amount Spent")
plt.xlabel("Amount Spent")
plt.show()
```

```python

# Full pairwise relationships
sns.pairplot(df)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
# Linear regression line for Length of Membership
sns.lmplot(data=df, x='Length of Membership', y='Yearly Amount Spent')
plt.title("Length of Membership vs Yearly Spend")
plt.show()
```
